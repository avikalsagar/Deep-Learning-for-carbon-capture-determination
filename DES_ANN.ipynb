```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
import matplotlib.font_manager
plt.rcParams["font.family"] = "serif"
```

```python
tf.__version__
```

```python
dataset = pd.read_csv('Final.csv')
X = dataset.iloc[ : , :-1].values
y = dataset.iloc[ : , -1].values
```

```python
print(X)
```

```python
print(y)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X[:, 0] = le.fit_transform(X[:, 0])
```

```python
print(X)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X[:, 1] = le.fit_transform(X[:, 1])
```

```python
print(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.08, random_state = 0)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
print(y_train)
```

```python
print(y_test)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
ann = tf.keras.models.Sequential()
```

```python
ann.add(tf.keras.layers.Dense(units=13, activation='relu'))
```

```python
ann.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

```python
ann.compile(optimizer='adam', loss='mean_squared_error')
```

```python
history=ann.fit(X_train, y_train, batch_size = 11, epochs = 1000)
```

```python
print(ann.predict(sc.transform([[0, 1, 16, 36, 12, 36, 12, 116.0, 137.04, 1046.2, 298.15, 0.75]])))
```

```python
print(ann.predict(sc.transform([[0, 0, 16, 36, 12, 36, 12, 116.0, 152.69299999999998, 1008.0, 318.15, 1.25]])))
```

```python
print(ann.predict(sc.transform([[2, 0, 4, 12, 0, 12, 0, 19.1, 96.587, 1140.8, 298.15, 0.75]])))
```

```python
y_pred = ann.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
plt.plot(y_test, color = 'red', label = 'Real data')
plt.plot(y_pred, color = 'blue', label = 'Predicted data')
plt.xlabel('Serial number of data point in test set')
plt.ylabel('Solubility of $CO_2$ in mole fractions')
#plt.title('Prediction')
plt.legend()
plt.savefig("ANN_predictions",dpi=1500,bbox_inches = 'tight')
plt.show()
```
