```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.font_manager
plt.rcParams["font.family"] = "serif"
```

```python
dataset3 = pd.read_csv('Predicted values(2.5MPa).csv')
```

```python
dataset3 = dataset3.dropna()
dataset3
```

```python
plt.figure(figsize=(5, 5))
N = 4
ind = np.arange(N) 
width = 0.25
  
xvals = [0.103063, 0.131555, 0.197483, 0.190784]
bar1 = plt.bar(ind, xvals, width, color = 'r')
  
yvals = [0.089076, 0.114715, 0.171184, 0.165199]
bar2 = plt.bar(ind+width, yvals, width, color='g')
  
zvals = [0.063296, 0.088608, 0.138661, 0.141875]
bar3 = plt.bar(ind+width*2, zvals, width, color = 'b')
  
plt.xlabel("DES Mixtures")
plt.ylabel('$CO_2$ Solubility(in moles)',fontname="Times New Roman",color='black')
plt.title("Solubilty of $CO_2$ in DES mixtures at 2.5MPa",position=(0.5, 0.94),size=10)
  
plt.xticks(ind+width,['TMA-Cl(2:1)','TEA-Cl(2:1)','TBA-Cl(2:1)','TBA-Cl(3:1)'])
plt.legend( (bar1, bar2, bar3), ('298.15K', '308.15K', '318.15K'), bbox_to_anchor=(0.35,0.90) )
plt.savefig("3",dpi=1500,bbox_inches = 'tight')
plt.show()
```
