```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.font_manager
plt.rcParams["font.family"] = "serif"
```

```python
dataset = pd.read_csv('Economic Analysis.csv')
```

```python
dataset.head()
```

```python
fig = plt.figure(figsize =(6, 4))
sns.set_style("whitegrid")
fig, ax = plt.subplots(figsize=(10, 10))
box_plot=sns.boxplot(x = 'Type', y = 'Price(max 100 G)', data = dataset, width= 0.5)
plt.xlabel("Type of solvent", size=15)
plt.ylabel("Price of solvent (in INR)", fontname="Times New Roman",color='black',size=15)
plt.xticks(size= 12)
plt.title("Box Plot representation of the prices of various solvents used in carbon capture", position=(0.5, 0.97),size=12)
plt.savefig("Economic Analysis",dpi=1500,bbox_inches = 'tight')
plt.show()
```
